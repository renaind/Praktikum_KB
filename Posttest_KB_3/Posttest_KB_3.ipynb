```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('vegetable_market.csv')
```

GRAFIK GARIS : Jumlah Data Sayur Dalam Setahun 

```python
plt.plot(data['Vegetable'], data['Price per kg'])
plt.title('Data Sayur Per Kilogram Dalam Setahun')
plt.xticks(rotation = 80)
plt.xlabel("Nama Sayur")
plt.ylabel("Jumlah")
plt.show()
```

SCATTER PLOT : Data Musim Sayur Berdasarkan Bulan

```python
plt.scatter(data['Vegetable'], data['Month'])
plt.xticks(rotation = 80)
plt.title('Data Musim Sayur Berdasarkan Bulan')
plt.xlabel("Nama Sayur")
plt.ylabel("Bulan")
plt.show()
```

HISTOGRAM : Data Jumlah Sayuran

```python
plt.hist(data['Vegetable'], color = 'blue')
plt.xticks(rotation = 80)
plt.title('Data Sayur Per Kilogram Dalam Setahun')
plt.xlabel("Nama Sayur")
plt.ylabel("Jumlah")
plt.show()
```

Persentase Kerusakan Sayur Dalam 3 Bulan

```python
plt.pie(data['Deasaster Happen in last 3month'].value_counts(), colors = ['red', 'green'], labels=['Tidak', 'Ya'] ,startangle = 90, autopct = '%.1f%%')
plt.title('Kerusakan yang Terjadi dalam 3 bulan')
plt.show()
```
